financial data gathering

```python
import requests
from io import StringIO
headers = { "User-Agent": ("Mozilla/5.0 (Windows NT 10.0; Win64; x64; rv:125.0) ""Gecko/20100101 Firefox/125.0")}
url = "https://finance.yahoo.com/markets/world-indices/"
response = requests.get(url, headers=headers)
tables = pd.read_html(StringIO(response.text))
major_indices = tables
major_indices1 = major_indices[0]
major_indices1 = major_indices1.dropna(axis = 0, how="all")
major_indices1
```

```python
ticker_list= major_indices1["Symbol"].to_list()
ticker_data= yf.Tickers(ticker_list)
ticker_data
```

```python
df = yf.download(ticker_list, period="1d", start= "2024-01-13", end = "2025-06-02")
```

```python
df
df.style.set_table_attributes('style="overflow-x: auto; display: block;"')
df.columns
```

```python
adj_close = df.dropna(thresh=10, axis=1)['Close']

adj_close.describe().T
```

```python
adj_close.plot(figsize=(22,36),subplots= True);
```
